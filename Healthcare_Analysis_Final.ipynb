### Analytical Methods

- **Provider-to-Population Ratio:** Primary care physicians per 100,000 residents
- **Shortage Indicator:** Counties below HRSA threshold of 60 providers per 100,000
- **Healthcare Access Score:** Composite measure (1-5 scale)
- **Statistical Techniques:** Descriptive statistics, correlation analysis, t-tests

**Primary Metrics Definitions:** 
- **Provider-to-Population Ratio:** Calculated as (Primary Care Physicians / County Population) × 100,000, this metric standardizes provider availability relative to population size, enabling valid comparisons across counties of different sizes.
- **Shortage Indicator:** Binary classification based on the HRSA threshold - counties with fewer than 60 primary care providers per 100,000 residents are designated as shortage areas.
- **Healthcare Access Score:** A composite measure on a 1-5 scale combining provider density quintiles and CDC ACCESS2 data. Counties scoring ≤2 are classified as "healthcare deserts" representing severely limited access.
- **Population Classification:** Counties are classified as low population (<50,000 residents) or high population (≥50,000) to enable urban-rural comparison, as the 50,000 threshold represents a common demarcation in rural definitions.

**Statistical Techniques:**
  - **Descriptive Analysis:** Calculation of means, medians, standard deviations, and distributions to characterize provider availability patterns across geographic areas and socioeconomic strata.
  - **Correlation Analysis:** Pearson correlation coefficients quantify linear relationships between provider density and socioeconomic variables, with values ranging from -1 (perfect negative relationship) to +1 (perfect positive relationship).
  - **Group Comparisons:** Independent samples t-tests assess whether observed differences between groups (e.g., high vs. low population counties) are statistically significant or could have occurred by chance.
  - **Stratified Analysis:** Counties are subdivided by poverty quartiles and population size to examine how access varies across socioeconomic strata and to identify interaction effects where multiple disadvantages compound.

### Study Scope and Boundaries
- This analysis focuses specifically on primary care physician availability at the county level across the contiguous United States. The scope deliberately excludes specialist physician shortages (which follow different geographic patterns), nurse practitioners and physician assistants (alternative provider types), telehealth availability (emerging modality not captured in traditional provider counts), and individual-level barriers to access beyond geographic availability (such as insurance status, transportation, cultural factors).
- The county-level aggregation necessarily obscures within-county variation - some counties contain both well-served and underserved communities. The analysis examines associations and correlations but cannot establish causal relationships given the cross-sectional design.

### Data Cleaning & Merging

The following code performs:
1. **FIPS Code Standardization:** Converts all county identifiers to standardized 5-digit FIPS codes with leading zeros
2. **Variable Extraction:** Selects relevant variables from each dataset (provider counts, population, poverty rates, income, access measures)
3. **Data Pivoting:** Transforms PLACES and County Health Rankings from long format (multiple rows per county) to wide format (one row per county)
4. **Multi-Source Integration:** Merges all datasets using FIPS codes as the common key
5. **Derived Metric Calculation:** Computes provider ratios, shortage indicators, and access scores


```python
# 1. CDC PLACES DATA
places_columns = {
    'fips': 'LocationID',           # County FIPS code
    'state': 'StateDesc',            # State name
    'county': 'LocationName',        # County name
    'measure': 'MeasureId',          # Health measure identifier
    'value': 'Data_Value'            # The actual data value
}
places_measures = [
    'ACCESS2',    # Access to healthcare (PRIMARY)
    'CHECKUP'     # Annual checkup
]

# 2. HRSA AHRF DATA (Health Professionals + Population)
ahrf_hp_columns = {
    'fips': 'fips_st_cnty',
    'county_name': 'cnty_name_st_abbrev',
    'pcp_count': 'phys_nf_prim_care_pc_exc_rsdt_22'  # Primary care physicians 2022
}
ahrf_pop_columns = {
    'fips': 'fips_st_cnty',
    'population': 'popn_22'  # Population 2022
}

# 3. COUNTY HEALTH RANKINGS DATA
chr_columns = {
    'fips': 'FIPS_Code',
    'state': 'Stabr',
    'county_name': 'Area_Name',
    'attribute': 'Attribute',
    'value': 'Value'
}
chr_attributes = {
    'MEDHHINC_2023': 'MedianIncome',           # Median household income
    'PCTPOVALL_2023': 'PovertyRate',           # Poverty rate %
    'Rural_Urban_Continuum_Code_2023': 'RuralUrbanCode'  # Rural-urban classification
}

# Load PLACES
places = pd.read_csv('data/county_data.csv')

# Load AHRF (both files)
ahrf_hp = pd.read_csv('data/ahrf2024hp.csv')
ahrf_pop = pd.read_csv('data/ahrf2024pop.csv')

# Load CHR
chr = pd.read_csv('data/economic_data.csv')

# Standardize FIPS (add leading zeros)
places['FIPS'] = places['LocationID'].astype(str).str.zfill(5)
places_filtered = places[places['MeasureId'].isin(places_measures)].copy()

# Pivot to wide format (one row per county)
places_pivot = places_filtered.pivot_table(
    index='FIPS',
    columns='MeasureId',
    values='Data_Value',
    aggfunc='first'
).reset_index()

# Add state and county names
location_info = places[['LocationID', 'StateDesc', 'LocationName']].drop_duplicates()
location_info['FIPS'] = location_info['LocationID'].astype(str).str.zfill(5)
places_clean = places_pivot.merge(location_info[['FIPS', 'StateDesc', 'LocationName']], 
                                   on='FIPS', how='left')

# Clean FIPS codes
ahrf_hp['FIPS'] = ahrf_hp['fips_st_cnty'].astype(str).str.zfill(5)
ahrf_pop['FIPS'] = ahrf_pop['fips_st_cnty'].astype(str).str.zfill(5)

# Select needed columns from health professionals
ahrf_hp_clean = ahrf_hp[['FIPS', 'phys_nf_prim_care_pc_exc_rsdt_22']].copy()
ahrf_hp_clean = ahrf_hp_clean.rename(columns={
    'phys_nf_prim_care_pc_exc_rsdt_22': 'PCP_count'
})

ahrf_pop_clean = ahrf_pop[['FIPS', 'popn_22']].copy()
ahrf_pop_clean = ahrf_pop_clean.rename(columns={
    'popn_22': 'Population'
})

# Merge AHRF files together
ahrf_merged = ahrf_hp_clean.merge(ahrf_pop_clean, on='FIPS', how='outer')

chr['FIPS'] = chr['FIPS_Code'].astype(str).str.zfill(5)

chr_filtered = chr[chr['Attribute'].isin(chr_attributes.keys())].copy()

chr_pivot = chr_filtered.pivot_table(
    index='FIPS',
    columns='Attribute',
    values='Value',
    aggfunc='first'
).reset_index()

# Rename columns to meaningful names
rename_dict = {attr: name for attr, name in chr_attributes.items()}
chr_clean = chr_pivot.rename(columns=rename_dict)

df = places_clean.copy()

# Merge AHRF
df = df.merge(ahrf_merged, on='FIPS', how='left', suffixes=('', '_ahrf'))

# Merge CHR
df = df.merge(chr_clean, on='FIPS', how='left', suffixes=('', '_chr'))

# Provider per 100k population
if 'PCP_count' in df.columns and 'Population' in df.columns:
    df['PCP_per_100k'] = (df['PCP_count'] / df['Population']) * 100000
    df['PCP_per_100k'] = df['PCP_per_100k'].clip(upper=300)

# Shortage indicator
if 'PCP_per_100k' in df.columns:
    df['is_shortage'] = df['PCP_per_100k'] < 60

# Low population indicator (proxy for rural)
if 'Population' in df.columns:
    df['is_low_population'] = df['Population'] < 50000

df['access_score'] = 0
components = 0

if 'PCP_per_100k' in df.columns:
    df['pcp_quintile'] = pd.qcut(df['PCP_per_100k'].fillna(0), 
                                  q=5, labels=[1,2,3,4,5], duplicates='drop')
    df['access_score'] += df['pcp_quintile'].astype(float)
    components += 1
    
if 'ACCESS2' in df.columns:
    # ACCESS2 is % lacking access - lower is better, so invert
    df['access2_quintile'] = pd.qcut(df['ACCESS2'].fillna(100), 
                                      q=5, labels=[5,4,3,2,1], duplicates='drop')
    df['access_score'] += df['access2_quintile'].astype(float)
    components += 1
    
if components > 0:
    df['access_score'] = df['access_score'] / components
    
if 'access_score' in df.columns:
    df['is_healthcare_desert'] = df['access_score'] <= 2

key_cols = ['FIPS', 'StateDesc', 'LocationName', 'PCP_count', 'Population', 
            'PCP_per_100k', 'MedianIncome', 'PovertyRate', 'ACCESS2']
for col in key_cols:
    if col in df.columns:
        missing = df[col].isnull().sum()
        pct = (missing / len(df)) * 100
        print(f"  {col}: {missing} ({pct:.1f}%)")

df.to_csv('data/merged_healthcare_data.csv', index=False)
print(f"Saved to: data/merged_healthcare_data.csv")
```

## Research Question 2: Healthcare Desert Identification

**Question:** Which counties can be classified as "healthcare deserts" with severely limited access?

### Analytical Approach:
This analysis employs a composite access score combining provider density and CDC access measures to identify counties with severely limited healthcare access. Counties scoring in the bottom two quintiles (access score ≤2) are designated as "healthcare deserts." which can then profile these deserts across multiple dimensions such as: provider availability, socioeconomic characteristics, population size. From there I identify the specific most underserved counties. The analysis includes statistical comparison between desert and non-desert counties to quantify the magnitude of differences.

## Research Question 3: Socioeconomic Correlations

**Question:** How do healthcare shortages correlate with socioeconomic factors?

### Analytical Approach

This analysis examines the strength and direction of linear relationships between provider availability and key socioeconomic indicators using Pearson correlation coefficients. We calculate a correlation matrix that covers provider density, poverty rates, median income, population size, and the composite access score. The analysis quantifies the degree that these variables move together, identifies the strongest predictors of provider availability, and displays these relationships through scatter plots. Statistical significance testing is how we can determine whether observed correlations exceed what would be expected by chance.

### Results

```python
def research_question_3(df):
    # Select variables for correlation
    var_mapping = {
        'PCP_per_100k': 'Provider Density',
        'PovertyRate': 'Poverty Rate',
        'MedianIncome': 'Median Income',
        'Population': 'Population',
        'access_score': 'Access Score'
    }
    
    available_vars = [v for v in var_mapping.keys() if v in df.columns and df[v].notna().sum() > 100]
    
    if len(available_vars) < 2:
        print("Insufficient data for correlation analysis")
        return None
    
    # Calculate correlation matrix
    corr_df = df[available_vars].copy()
    corr_matrix = corr_df.corr()
    
    print("Correlation Matrix:")
    print(corr_matrix.round(3))
    print()
    
    # Key findings
    print("Key Correlation Findings:")
    if 'PCP_per_100k' in corr_matrix.columns:
        pcp_corrs = corr_matrix['PCP_per_100k'].drop('PCP_per_100k').sort_values(key=abs, ascending=False)
        for var, corr_val in pcp_corrs.items():
            strength = 'strong' if abs(corr_val) > 0.5 else 'moderate' if abs(corr_val) > 0.3 else 'weak'
            direction = 'positive' if corr_val > 0 else 'negative'
            print(f"  Provider Density ↔ {var_mapping[var]}: r = {corr_val:.3f} ({strength} {direction})")
    print()
    
corr_matrix = research_question_3(df)
```

## Research Question 4: Social Determinants Deep Dive

**Question:** What is the detailed relationship between access and poverty/population factors?

### Analytical Approach
This analysis stratifies counties into poverty quartiles and population size categories to examine how multiple disadvantages combine. We calculate provider density within each stratum and use ANOVA and t-tests to assess statistical significance. The interaction analysis examines whether poverty and population effects compound or simply add together.

### Results

```python
def research_question_4(df):
    results = {}
    # Poverty level analysis
    if 'PovertyRate' in df.columns and 'PCP_per_100k' in df.columns:
        print("Analysis by Poverty Level:")
        df['poverty_quartile'] = pd.qcut(df['PovertyRate'].fillna(df['PovertyRate'].median()), 
                                         q=4, labels=['Low', 'Medium-Low', 'Medium-High', 'High'],
                                         duplicates='drop')
        
        poverty_analysis = df.groupby('poverty_quartile', observed=True).agg({
            'PCP_per_100k': ['mean', 'median', 'std'],
            'is_shortage': 'sum',
            'FIPS': 'count'
        }).round(2)
        
        poverty_analysis.columns = ['Avg_PCP', 'Median_PCP', 'StdDev', 'Shortage_Count', 'Total_Counties']
        poverty_analysis['Shortage_Pct'] = (poverty_analysis['Shortage_Count'] / 
                                           poverty_analysis['Total_Counties'] * 100).round(1)
        
        print(poverty_analysis)
        print()
        results['poverty_analysis'] = poverty_analysis
    
    # Population size analysis
    if 'is_low_population' in df.columns and 'PCP_per_100k' in df.columns:
        print("Population Size Comparison:")
        pop_comparison = df.groupby('is_low_population').agg({
            'PCP_per_100k': ['mean', 'median'],
            'PovertyRate': 'mean' if 'PovertyRate' in df.columns else lambda x: np.nan,
            'MedianIncome': 'mean' if 'MedianIncome' in df.columns else lambda x: np.nan,
            'is_shortage': lambda x: (x.sum() / len(x) * 100)
        }).round(2)
        
        pop_comparison.columns = ['Avg_PCP', 'Median_PCP', 'Avg_Poverty', 'Avg_Income', 'Shortage_Pct']
        pop_comparison.index = ['High Population (>50k)', 'Low Population (<50k)']
        print(pop_comparison)
        print()
        results['population_comparison'] = pop_comparison
        
        # Statistical test
        low_pop = df[df['is_low_population']==True]['PCP_per_100k'].dropna()
        high_pop = df[df['is_low_population']==False]['PCP_per_100k'].dropna()
        
        if len(low_pop) > 0 and len(high_pop) > 0:
            t_stat, p_value = stats.ttest_ind(low_pop, high_pop)
            print(f"Statistical Test - Low vs High Population:")
            print(f"  t-statistic: {t_stat:.4f}")
            print(f"  p-value: {p_value:.4e}")
            print(f"  Significant difference: {'Yes' if p_value < 0.05 else 'No'} (α=0.05)")
            print()

social_results = research_question_4(df)
```

## Risk Assessment and Limitations

- While our analysis provides robust insights, several limitations and risk factors must be acknowledged:

**Model Limitations:**
- Cross-Sectional Design: Our analysis examines a single time period, preventing assessment of temporal trends or definitive causal inference. We document associations but cannot definitively establish that poverty causes provider shortages versus shortages contributing to poverty.

- Provider Measurement: Analysis focuses on primary care physicians, excluding nurse practitioners, physician assistants, and other providers who contribute substantially to primary care delivery. **Policy and Implementation Risks:**
- Political Feasibility: Recommendations requiring substantial resource commitments or regulatory changes face political and budgetary constraints that may limit implementation.
